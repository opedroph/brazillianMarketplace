# 2 - Análises e tratamentos iniciais dos dados


```python
# toda a nossa base de dados está contida em diferentes datasets, vamos importá-los e depois juntá-los em uma grande base 
def merge_datasets(datasets):
    merged = pd.merge(datasets['olist_orders_dataset'], datasets['olist_customers_dataset'], on='customer_id', how='left')

    # um dicionário com o nome do dataframe e a chave que liga esse a outro dataframe 
    merge_keys = {
        'order_items': 'order_id',
        'order_payments': 'order_id',
        'order_reviews': 'order_id',
        'sellers': 'seller_id',
        'products': 'product_id',
        'product_category_name_translation': 'product_category_name'  # serve para traduzir os nomes de categorias, vamos usá-lo uma vez que toda
        # a base está em inglês
    }

    for name, key in merge_keys.items():
        if name in datasets:
            merged = pd.merge(merged, datasets[name], on=key, how='left')

    return merged


# carregando os datasets
olist_customers_df = pd.read_csv('olist_customers_dataset.csv')
order_items_df = pd.read_csv('olist_order_items_dataset.csv')
order_payments_df = pd.read_csv('olist_order_payments_dataset.csv')
order_reviews_df = pd.read_csv('olist_order_reviews_dataset.csv')
olist_orders_df = pd.read_csv('olist_orders_dataset.csv')
sellers_df = pd.read_csv('olist_sellers_dataset.csv')
products_df = pd.read_csv('olist_products_dataset.csv')
product_category_name_translation_df = pd.read_csv('product_category_name_translation.csv')

datasets = {
    'olist_customers_dataset': olist_customers_df,
    'order_items': order_items_df,
    'order_payments': order_payments_df,
    'order_reviews': order_reviews_df,
    'olist_orders_dataset': olist_orders_df,
    'sellers': sellers_df,
    'products': products_df,
    'product_category_name_translation': product_category_name_translation_df
}

#retornando o dataset construido e com todas as colunas
merged_df = merge_datasets(datasets)
```

## Engenharia de atributos
criando colunas que podem ser importantes para a análise de dados posterior

```python
#criando algumas colunas que podem ser importantes para análise

#tempo total para envio, do momento em que a compra é aprovada até chegar ao consumidor
merged_df['total_time_to_delivery'] = (merged_df['order_delivered_customer_date'] - merged_df['order_approved_at']).dt.days
#diferença de tempo entre a compra, e ela ser aprovada
merged_df['time_to_processing_order'] = (merged_df['order_approved_at'] - merged_df['order_purchase_timestamp']).dt.days
#tempo em que a empresa demora pra entregar a encomenda para o parceiro de logistica
merged_df["time_to_delivery_company"] = (merged_df['order_delivered_carrier_date'] - merged_df['order_approved_at']).dt.days
#tempo que a parceira de logistica demora pra entregar a encomenda para o cliente
merged_df["time_to_delivery_carrier"] = (merged_df['order_delivered_customer_date'] - merged_df['order_delivered_carrier_date']).dt.days
# diferença entre o dia estipulado e o real de entrega
merged_df['Days_late_delivery'] = (merged_df['order_delivered_customer_date'] - merged_df['order_estimated_delivery_date']).dt.days
#preço total da encomenda (produto + frete)
merged_df['total_order_price'] = merged_df['price'] + merged_df['freight_value']

# criando variável para dizer se a entrega foi ou nao atrasada
merged_df['late_delivery'] = (merged_df['Days_late_delivery'] > 0).astype(int)

#convertendo a satisfação em uma variavel binária
merged_df["binary_review"] = (merged_df["review_score"] >=4).astype(int)


# criando variáveis de tempo para análises
merged_df['order_month'] = merged_df['order_purchase_timestamp'].dt.month
merged_df['order_day_week'] = merged_df['order_purchase_timestamp'].dt.dayofweek
merged_df['order_hour'] = merged_df['order_purchase_timestamp'].dt.hour
merged_df['order_year'] = merged_df['order_purchase_timestamp'].dt.year
merged_df['order_day'] = merged_df['order_purchase_timestamp'].dt.day
```

## Analisando as entregas por diferentes períodos de tempo

## Usando Diagrama de Pareto e Infomation Value para definir os estados e produtos que mais vendem

```python
#organizando o gráfico para ficar os maiores valores no início
states_analysis.sort_values(by="entregas com atraso", inplace=True, ascending=False)

#calculando o percentual de entregas atrasadas por estado de forma cumulativa
total_late_deliveries = states_analysis["entregas com atraso"].sum()
total_late_deliveries
late_deliveries_by_state_percent = (states_analysis["entregas com atraso"]/ total_late_deliveries).cumsum() * 100

#colocando o gráfico com a quantidade de entregas realizadas com atraso e sua porcentagem de forma cumulativa
fig, ax = plt.subplots(figsize=(14, 9))
states_analysis["entregas com atraso"].plot(kind='bar', ax=ax, color='skyblue')
ax2 = ax.twinx()
ax2.plot(late_deliveries_by_state_percent.index, late_deliveries_by_state_percent.values, color='red', marker='D', ms=7)
ax2.yaxis.set_major_formatter(PercentFormatter())

#adicionando as linhas para identificar os 80% mais importantes
ax2.axhline(80, color='green', linestyle='--', linewidth=2)
states_80_idx = late_deliveries_by_state_percent[late_deliveries_by_state_percent >= 80].index[0]
states_80_position = late_deliveries_by_state_percent.index.get_loc(states_80_idx)

ax.axvline(states_80_position, color='purple', linestyle='--', linewidth=2)

plt.show()
```

* Uma boa estratégia seria focar nos estados que possuem muitas entregas atrasadas e uma porcentagem alta também, dessa forma impactando consideravelmente nas entregas atrasadas e melhorando a visão da empresa em alguns Estados específicos. 

```python
#separando o codigo postal do cliente em postal zones
#para infomrações: https://pt.wikipedia.org/wiki/Código_de_Endereçamento_Postal
merged_df["postal_zone"] = merged_df["customer_zip_code_prefix"].astype(str)
merged_df["postal_zone"] = merged_df["postal_zone"].apply(lambda x: x[0] if len(x)>4 else 0)
merged_df["postal_zone"] 
```

```python
# Loop
target = "late_delivery"
for i in merged_df[["postal_zone" ]]:
    
    df_woe_iv = pd.crosstab(merged_df[i], merged_df[target], normalize='columns')
    df_woe_iv['woe'] = np.log(df_woe_iv[1] / df_woe_iv[0])
    df_woe_iv['woe'].replace({-np.inf: 0, np.inf: 0}, inplace=True)  #serve para ignorar as categorias que contenham 0 valores para verdadeiro ou falso
    df_woe_iv['iv'] = (df_woe_iv['woe'] * (df_woe_iv[1] - df_woe_iv[0])).sum()
    
    df_woe_iv["count_0"] = merged_df.groupby(i)[target].apply(lambda x: (x == 0).sum())
    df_woe_iv["count_1"] = merged_df.groupby(i)[target].apply(lambda x: (x == 1).sum())
    df_woe_iv["count_total"] = df_woe_iv["count_0"] + df_woe_iv["count_1"]
    df_woe_iv["% ocorrências"] = (df_woe_iv["count_1"] / (df_woe_iv["count_total"])) * 100


    print(df_woe_iv.sort_values(by="count_total", ascending=False).to_string(max_rows=None, max_cols=None),'\n------------------------------------------------------------\n\n')
```

* Utilizando calculo de Weight of Evidence (WOE) e Information Value (IV), podemos definir o quão forte para previsão é uma variável

Interpretamos o resultado de IV assim:

    < 0.02, não deve ser usado para previsão
    0.02 - 0.1, preditor fraco
    0.1 - 0.3, preditor médio
    0.3 - 0.5, preditor forte
    > 0.5, parece bom demais para ser verdade

* Como podemos ver, o "postal_zone" é uma variável categorica com um poder preditivo moderado, podendo ser usados em algoritmos de machine learning para prever se uma encomenda será ou não entregue com atraso
*  Dessa forma, conseguimos notar que a postal zone impacta muito na ocorrência de atrasos
  

```python
late_deliveries_by_category = merged_df[merged_df['late_delivery'] == 1]['product_category_name_english'].value_counts()
total_late_deliveries = late_deliveries_by_category.sum()
late_deliveries_by_category_percent = (late_deliveries_by_category / total_late_deliveries).cumsum() * 100


fig, ax = plt.subplots(figsize=(14, 9))
late_deliveries_by_category.plot(kind='bar', ax=ax, color='skyblue')
ax2 = ax.twinx()
ax2.plot(late_deliveries_by_category_percent.index, late_deliveries_by_category_percent.values, color='red', marker='D', ms=7)
ax2.yaxis.set_major_formatter(PercentFormatter())

ax2.axhline(80, color='green', linestyle='--', linewidth=2)
category_80_idx = late_deliveries_by_category_percent[late_deliveries_by_category_percent >= 80].index[0]
category_80_position = late_deliveries_by_category.index.get_loc(category_80_idx)

ax.axvline(category_80_position, color='purple', linestyle='--', linewidth=2)

ax.tick_params(axis='y', colors='skyblue')
ax2.tick_params(axis='y', colors='red')
ax.set_xlabel('Product Category')
ax.set_ylabel('Number of Late Deliveries', color='skyblue')
ax2.set_ylabel('Cumulative Percentage', color='red')
plt.title('Pareto Chart of Late Deliveries by Product Category with 80/20 Threshold')

plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

* Aqui é possível notar que apenas 15 produtos constituem 80% das vendas desse e-comerce
* distribuir melhor esses produtos entre os estados que mais compram pode diminuir consideravelmente o tempo de entrega
